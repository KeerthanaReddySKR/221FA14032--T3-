```python
# prompt: to get better linear regression graph which coloumns should be selected give python code between TS1 and TS126

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Load the dataframe
df = pd.read_csv('/content/Facebook_Obama.csv')

# Initialize variables to store best fit
best_r2 = 0
best_x_col = None
best_y_col = None

# Iterate over all possible combinations of columns
for i in range(1, 127):
  for j in range(i + 1, 127):
    # Select the relevant columns
    X = df[['TS' + str(i)]]
    y = df['TS' + str(j)]

    # Create a linear regression model
    model = LinearRegression()

    # Fit the model to the data
    model.fit(X, y)

    # Make predictions
    y_pred = model.predict(X)

    # Calculate R-squared
    r2 = r2_score(y, y_pred)

    # Update best fit if current fit is better
    if r2 > best_r2:
      best_r2 = r2
      best_x_col = 'TS' + str(i)
      best_y_col = 'TS' + str(j)

# Print the best fit
print(f"Best fit found with {best_x_col} as X and {best_y_col} as Y, R-squared: {best_r2}")

# Select the best columns
X = df[[best_x_col]]
y = df[best_y_col]

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)

# Plot the data and the regression line
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red')
plt.xlabel(best_x_col)
plt.ylabel(best_y_col)
plt.title(f'Linear Regression of {best_y_col} vs {best_x_col}')
plt.show()

# Print the model coefficients
print('Intercept:', model.intercept_)
print('Slope:', model.coef_[0])
```

```python
# prompt: to get better linear regression graph which coloumns should be selected give python code between TS1 and TS126

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Load the dataframe
df = pd.read_csv('/content/LinkedIn_Economy.csv')

# Initialize variables to store best fit
best_r2 = 0
best_x_col = None
best_y_col = None

# Iterate over all possible combinations of columns
for x_col in range(1, 127):
  for y_col in range(1, 127):
    if x_col != y_col:
      # Select the relevant columns
      X = df[['TS{}'.format(x_col)]]
      y = df['TS{}'.format(y_col)]

      # Create a linear regression model
      model = LinearRegression()

      # Fit the model to the data
      model.fit(X, y)

      # Make predictions
      y_pred = model.predict(X)

      # Calculate R-squared
      r2 = r2_score(y, y_pred)

      # Update best fit if current fit is better
      if r2 > best_r2:
        best_r2 = r2
        best_x_col = 'TS{}'.format(x_col)
        best_y_col = 'TS{}'.format(y_col)

# Print the best fit
print('Best fit:')
print('X column:', best_x_col)
print('Y column:', best_y_col)
print('R-squared:', best_r2)

# Select the best columns
X = df[[best_x_col]]
y = df[best_y_col]

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)

# Plot the data and the regression line
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red')
plt.xlabel(best_x_col)
plt.ylabel(best_y_col)
plt.title('Linear Regression of {} vs {}'.format(best_y_col, best_x_col))
plt.show()

# Print the model coefficients
print('Intercept:', model.intercept_)
print('Slope:', model.coef_[0])
```
